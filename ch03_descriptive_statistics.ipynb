# 3. Descriptive statistics of sample data

In this chapter, we discuss the most important **statistics for location, variation and skewness** of a sample set. The number of statistics that can be calculated depends on the nature of the data. When statistics are calculated for an entire population or process, they are called **parameters**. Statistics are represented with Roman letters ($a,b,c$) while parameters use Greek letters ($\alpha, \beta, \gamma$). The to-be discussed statistics are tabulated below for the various data types.

Type                             | Nominal          | Ordinal                                              | Interval
---------------------------------|------------------|------------------------------------------------------|---------------------------------------------------------------------
Location                         | Mode             | Mode, median, quartiles                              | Mode, median, quartiles, arithmetic mean, geometric mean
Variation                        | Dispersion index | Range, interquartile range, dispersion index         | Range, interquartile range, mean absolute deviation, variance and standard deviation, coefficient of variation
Moments                          |                  |                                                      | (Non-)central moment
Skewness                         |                  |                                                      | Pearson, Fisher
Steepness                        |                  |                                                      | Kurtosis
Location, variation and skewness |                  | Box plot                                             | Box plot
Correlation or association       |                  | Rank correlation coefficient                         | Covariance, correlation coefficient, rank correlation coefficient

Throughout, we denote the number of observations in a sample, or the **sample size** by $n$. Whenever we discuss a quantitative variable, we call the variable $x$ and $n$ observations for that variable $x_1, x_2, ..., x_n$. Whenever we discuss two variables, we refer to them as $x$ and $y$, and denote the $n$ observations by $x_1, x_2, ..., x_n$ and $y_1, y_2, ..., y_n$.


## 3.1. Measures of central tendency or location

Location statistics are values that best describe the central tendency of the data. In other words, they give an indication of **how big or small the dataset is**.

### 3.1.1. Median

The **median $M_e$**  is the middle element of the orderedd dataset, so:

- if $n$ is odd, $M_e$ is the $\frac{n+1}{2}$-the element.
- if $n$ is even, $M_e$ is the mean of the $\frac{n}{2}$-th and the $\frac{n}{2} + 1$-th element.

Some properties of the median are:

- About $50\%$ of the observations are below or above the median.
- The sum of the absolute deviations of the observations $x_i$ from a constant c, $\sum_{i=1}^n |x_i - c|$, is minimal if $c = M_e$.
- The median is not affected by a few extremely large or extremely small observations.
- The median is the mean of a truncated data set in which only the middle (two middle) observation(s) remain(s).

### 3.1.2. Mode

The **mode $M_o$** of a sample is the observation with the highest frequency. The mode makes most sense as a statistic for **nominal, ordinal, interval and rational variables**. For continuous quantitative variables, better statistics are most often available. The mode may not be unique; it sometimes happens that there are several values or classes with the same highest frequency. This is reflected in the histogram as well; when two or more peaks occur, one speaks of a **bimodal or multimodal distribution**. If a histogram has only one peak, it is called **unimodal**.

### 3.1.3. Arithmetic mean

The **arithmetic mean** or sample mean $x$ of $n$ observations $x_1, x_2, ..., x_n$ is $$x = \frac{1}{n}\sum_{i=1}^{n} x_i.$$
For **grouped data**, this becomes: $$x=\frac{1}{n}\sum_{i=1}^{k} f_i x_i.
$$

For a **finite population** with *N* elements, a slightly different notation is used: $$\mu = \frac{1}{N} \sum_{i=1}^{N} x_i.$$ However, in practice, **population mean $\mu$ is almost always unknown**, meaning we usually have to work with the arithemtic means $x$ of samples.

Some properties of the arithmetic mean are:

- The sum of all observations is equal to the arithmetic mean multiplied by the sample size: $\sum_{i=1}^{n} x_i = nx$.
- The sum of the deviations of the observations from the mean is zero: $\sum_{i=1}^{n} (x_i - x) = 0$.
- The sum of the squared deviations from a constant $c$, $\sum_{i=1}^{n} (x_i - c)^2$, is minimal if $c=x$.
- The arithmetic mean of a sample of constant values $a, ..., a$ equals the constant value itself: $x=a$.
- The arithmetic mean of a number of observations $ax_1 + b, ..., ax_n + b$ (where $a$ and $b$ are constants), obtained by linearly transforming an original set of observations, $x_1, ..., x_n$, can be obtained by applyng the linear transformation to the original mean $ax + b$.

### 3.1.4. Geometric mean

The **geometric mean $G$** of a set of observations $x_1, ..., x_n$ is: $$G = \left(\prod_{i=1}^{n} x_i\right)^{\frac{1}{n}}.$$ This definition implies that the geometric mean can only be caulcated for strictly positive observations. The geometric mean is always smaller or equal to the arithmetic mean. Both are equal only if all observations in a dataset are equal. Geometric mean often makes more sense in data sets that grow exponentially, like compound interest.


## 3.2. Measures of relative location

A measure of relative location indicates the position of an observation in comparison to the values of the other observations.

### 3.2.1. Order statistics, quantiles, percentiles, deciles

The **$i$-th order statistic $x_{(i)}$** in a sample of $n$ observations is the $i$-th observation after ranking the observations from small to large. The first-order statistic is the **minimum**, while the last-order statistic is the **maximum**. We call these two values $x_\textrm{min}$ and $x_\textrm{max}$.

The **$(100 \cdot p)$-th percentile or quantile $c_p$** of a sample, where $0<p<1$, is a real number that is greater than (about) $100 \cdot p \%$ of the observations and smaller than (about) $100 \cdot (1 - p) \%$ of the observations. A percentile can be calculated as follows:

- Sort the $n$ observations from small to large in positions $1, 2, ..., n$.
- Calculate the position of the $(100 \cdot p)$-th percentile as $q = p(n+1)$.
    - If $q$ is an integer, then $c_p = x_{(q)}$.
    - If $q$ is not an integer,
        - First, round $q$ down to the nearest integer, $a$.
        - Next, determine the difference, $f = q-a$.
        - Then, $c_p = (1-f) x_{(a)} + f x_{(a+1)} = x_{(a)} + f \left( x_{(a+1)} - x_{(x)} \right)$.

### 3.2.2. Quartiles

The **first, second and third quartiles $Q_1, Q_2, Q_3$** are the $25$-th, $50$-th and $75$-th percentile of the sample. In other words:
$$
Q_1 = c_{0.25}, \hspace{1cm} Q_2 = c_{0.5} = M_e, \hspace{1cm} Q_3 = c_{0.75}.
$$ A such, these three quartiles are generally shown, together with the arithmetic mean, in a so-called **boxplot**.

## 3.3. Measures of variation or spread

### 3.3.1. Range

The **range $R$** of a set of observations is the difference between the value of the largest and smallest observation: $$R = x_\mathrm{max} - x_\mathrm{min}.$$ Note that this is a very simplistic statistic, which is very sensitive to extreme values. In industry, $R$ if often used in **statistical process control**.

### 3.3.2. Interquartile range

The **interquartile range $Q$** is defined as the difference between the third and first quartile:
$$
Q = Q_3 - Q_1.
$$
Since half of the data is between $Q_1$ and $Q_3$, the itnerquartile range is a measure of spread for half of the data set. This measure of spread is insensitive to extreme values, as long as **less then $25 \%$ of the data values are extremely small and less than $25 \%$ are extremely large**.

### 3.3.3. Mean absolute deviation

The **mean absolute deviation $\mathrm{MAD}$**, is the arithmetic mean of the absolute values of the deviations of the observations from the arithmetic mean: $$\mathrm{MAD} = \frac{1}{n}\sum_{i=1}^{n} |x_i - x|.$$ Just like the arithmetic mean itself, the $\mathrm{MAD}$ is sensitive to extreme values.

### 3.3.4. Variance

The **sample variance $s^2$** of a set of $n$ observations $x_1, x_2, ..., x_n$ is: $$s^2 = \frac{1}{n-1}\left( \sum_{i=1}^{n} x_i^2 - n x^2 \right) = \frac{1}{n-1} \left\{
        \sum_{i=1}^{n} x_i^2 - \frac{1}{n} \left( \sum_{i=1}^{n} x_i \right)^2 
    \right\},$$

with the arithmetic mean: $$x = \frac{1}{n} \sum_{i=1}^{n} x_i.$$

If we would compute the variance of all $N$ elements of a population or process, we speak of **population variance** or **process variance**: $$\sigma^2 = \frac{1}{N} \sum_{i=1}^{N} (x_i - \mu)^2.$$

Note the difference in the used denominator: for the population of size $N$ the (more logical) $N$ is used, while for a sample of size $n$, we divide by $n-1$! The number $n-1$ is called the **degrees of freedom $\mathrm{DOF}$** of the data set, which we can imagine as the total collection of **information units**. In a sample of $n$ observations, we have $n$ units of information. Every statistical operation on this sample costs units of information, as it abstracts further from the measurement. Therefore, consecutive operations need to take this into account. In order to calculate $s^2$, $x$ needs to be calculated first. As the computation of $x$ requires $1\ \mathrm{DOF}$, we only have $n-1$ degrees of freedom left to calculate $s^2$ with.

To make more sense of this, imagine an experiment in which we are monitoring the growth rate $200$ cabbages in response to a lamp being on or off. For both sub-experiments of both states of the lamp, we will dedicate $n=100$ cabbages, of which the growth will be measured daily. For both states of the lamp, the growth of every cabbage will be measured against the mean cabbage. However, as long as we do not have an infinte number of cabbages to monitor, that *exactly average* cabbage does (likely) not exist. For example, on average, the cabbages might grow $1.253 \pm 0.154 \mathrm{cm/day}$, while none of the unique cabbages might grow that *exact* amount on any given day. To correct for this inexactness, it makes more sense to overestimate the variance by dividing by a smaller number i.e., the $\mathrm{DOF} = n - 1$, instead of a larger one i.e., $n$.

### 3.3.5. Standard deviation

The **sample standard deviation** is simply the square root of the variance: $$s = \sqrt{s^2}.$$ Conversely, the **population standar deviation $\sigma$** is the square root of the population variance: $$\sigma = \sqrt{\sigma^2}.$$ Note that for a sample of only 2 observations the range and sample standard deviation contain the same information.

### 3.3.6. Coefficient of variation

While variance and standard deviation are important, they are sometimes not the best choices as measures of spread. Consider 2 samples of 8 observations:

- Sample 1: $15, 20, 20, 30, 35, 35, 40, 45$
- Sample 2: $1015, 1020, 1020, 1030, 1035, 1035, 1040, 1045$

For both samples we find the same variance ($s = 114.2857$), while their means are drastically different ($x_1 = 30, x_2 = 1030$). Therefore, the relative variability of the first sample seems considerably larger than for the second one. For that reason, we can define a more robust estimator.

The **coefficient of variation $\mathrm{CV}$** is the ratio of the sample standard deviation and the arithmetic mean: $$\mathrm{CV} = \frac{s}{x}.$$ Because of this correction, $\mathrm{CV}$ is useful for comparing spread of data with different means and essential when comparing spread of data with different dimensions i.e., with different measurement units.

### 3.3.7. Dispersion indices for nominal and ordinal variables

The spread of nominal data (unordered data e.g., gender, nationality) measures **heterogeneity**, while it measures **polarization** for ordinal data (qualitatively ordered data e.g., customer satisfaction, climbing route difficulty).

#### 3.3.7.1. Nominal dispersion index

The **nominal dispersion index $\mathrm{nDi}$** of a nominal variable with $k$ different categories is based on the definition of $\mathrm{homogeneity} = 1 - \frac{1}{k}$, which determines **how well the observations fall within a single category** i.e., how homogeneous any given class is. Thus, for $k=1$, the maximal $\mathrm{homogeneity} = 0$, as all samples, regardless of their value, must belong to one and the same class. The more classes we add, the better any observation will fit with at least one of these more and more precise classes. However, there are diminishing returns as $\lim_{k \to \infty} \mathrm{homogeneity} \to 1$.

We can now define the nominal dispersion index as:
$$\mathrm{nDi} 
= 1 - \frac{\textrm{observed homogeneity}}{\textrm{maximal homogeneity}}
= 1 - \frac{
    \sum_{i=1}^k \left( f_i^* - \frac{1}{k} \right)^2
}{1 - \frac{1}{k}},$$

with $f_i^*$ the relative frequency of the $i$-th class.

The nominal dispersion index ranges $0 < \mathrm{nDi} < 1$. In case it is 0, there is no spread at all and all observations belong to the same category. Conversely, if it is 1, all observations are evenly distributed over all classes and heterogeneity is maximized.

#### 3.3.7.2. Ordinal dispersion index

The **ordinal dispersion index $\mathrm{oDi}$** of an ordinal variable with $k$ different categories is based on the definition of the **sum of squares**:
$$\sum_{i=1}^{k-1} \left( F_i^* - \frac{1}{2} \right)^2,$$
with $F_i^*$ the relative cumulative frequency of te $i$-th class. This sum of squares is a measure of concentration; the larger this measure, the larger the extent to which the observations fall within a single category. If concentration is minimal, the observations are distributed amongst a large number of classes.

The ordinal dispersion index is then: 
$$ \mathrm{oDi} 
= 1 - \frac{{\textrm{observed concentration}}}{\textrm{maximal concentration}}
= 1 - \frac{4}{k-1} \sum_{i=1}^{k-1} \left( F_i^* - \frac{1}{2} \right)^2,
$$
which always lies $0 < \mathrm{oDi} < 1$. In case it is 0, there is no polarization, all observations belong to the same class and the concentration is maximal. Conversely, if $\mathrm{oDi} = 1$, then polarization is maximal, all observations are evenly distributed between the two extreme classes.

## 3.4. Measures of skewness

Histograms can be either symmetric or asymmetric. The latter case is often called **skewed**. We can make two further distinctions: a **positively skewed** histogram is skewed to the right i.e., the positive right-hand tail is longer than the left-hand tail. Conversely, a **negatively skewed** histogram is skewed to the left i.e., the negative left-hand tail is longer than the right-hand tail.

In a simple unimodal histogram, skewness can be determined based on the positions of the **arithmetic mean, median and mode**. In a perfectly symmetrical histogram, the three statistics are identical. When a population becomes skewed, all three estmimators will shift towards the longer tail. The mean $x$ will move the furthest from, as it is most sensitive to outliers. The median $M_e$ will move less, and the mode $M_o$ will barely move. In case of very strong skewness, $M_o$ will even shift back to the same location as in a symmetrical distribution.

```python
# Constants
distributions = (sp.stats.norm, sp.stats.skewnorm)
skewness = 5
loc = 0
scale = 1
x = np.linspace(-5*scale, 5*scale, 1001)

# Creating three distributions.
norm = {'distr': sp.stats.norm(loc=loc, scale=scale), 
        'name': 'Normal distribution'}
skewnorm_positive = {'distr': sp.stats.skewnorm(loc=loc, scale=scale, a=skewness),
                     'name': 'Positive skewnormal distribution'}
skewnorm_negative = {'distr': sp.stats.skewnorm(loc=loc, scale=scale, a=-skewness),
                     'name': 'Negative skewnormal distribution'}

for distribution in (norm, skewnorm_negative, skewnorm_positive):
    # Grabbing attributes.
    distr = distribution['distr']
    pdf = distr.pdf(x)
    pdf_max = np.max(pdf)
    # Plotting the distribution.
    f, a = plt.subplots()
    plt.plot(x, pdf)

    # Finding the mean, median and mode.
    mean = distr.mean()
    median = distr.median()
    mode = x[np.argmax(pdf)]
    pearson = 3/distr.std() * (mean - median)

    # Plotting the mean, median and mode.
    a.axvline(mean, color=(1,0,0,0.33), label=f'Mean: {mean:.2f}')
    a.axvline(median, color=(0,1,0,0.33), label=f'Median: {median:.2f}')
    a.axvline(mode, color=(0,0,1,0.33), label=f'Mode: {mode:.2f}')
    
    # Decorating.
    a.set(xlabel='x', ylabel='Probability', title=f"{distribution['name']} $[S_P = {pearson:.2f}]$")
    a.legend()
```

As all of the above estimators ($x, M_e, M_o$) are incapable of adequately describing skewness on their own, **Pearson's coefficient of skewness $S_P$** is typically used instead, which is defined based on 3 other estimators as follows:
$$
S_P = \frac{3}{s} (x - M_e),
$$
with $s$ the sample standard deviation, $x$ the sample mean and $M_e$ the sample median. Note from the above plots that skewness can indeed be measured by taking the difference in shifts of the mean and median. However, this difference could be made arbitrarily small or large by changing the measurement unit. This effect can be compensated by simply dividing by the standard deviation, which is always in the same units as the measurement. As the mean $x$ always shifts more than the median $M_e$, Pearson's coefficient for skewness $S_P > 0$ for a right-skewed (positively skewed) distribution, and $S_P < 0$ for a left-skewed (negatively skewed) distribution.

Alternatively, skewness can be calculated from the **third central moment $m_3$** of the sampled data. Generally, the $k$-the central moment of a sample of size $n$ is defined as:
$$
m_k = \frac{1}{n} \sum_{i=1}^{n} (x_i - x)^k.
$$

Finally, the **skewness of Fisher $S_F$** is perhaps the most commonly used measure for skewness, based on a variation of $\frac{m_3}{s^3}$. A typical definition is:
$$
S_F = \frac{n^2}{(n-1)(n-2)} \frac{m_3}{s^3}
    = \frac{n^2}{(n-1)(n-2)} \sum_{i=1}^{n} \left( \frac{x_i - x}{s} \right)^3.
$$
Fisher's skewness is a dimensionless number which is 0 for a symmetric distribution ($S_F = 0$), positive for a positively skewed distribution ($S_F > 0$) and negative for a negatively skewed distribution ($S_F < 0$).

## 3.5. Kurtosis

The extend to which a **distribution has a sharp peak** is quantified by **kurtosis $g$**, which can be viewed as a measure for steepness. The kurtosis can be described as:
$$
g = \frac{n (n+1)}{(n-1)(n-2)(n-3)} \sum_{i=1}^{n} \left( \frac{x_i - x}{s} \right)^4 - \frac{3 (n-1)^2}{(n-2)(n-3)}.
$$
Similarly to skewness, kurtosis is a dimensionsless number and is zero for normally distributed data. In case the distribution has a sharper peak than the normal distribution, $g > 0$. Conversly, when the distribution is flatter than the normal distribution, $g < 0$.

## 3.6. Transformation and standardization of data

Let us study the effects of a linear transformation $y_i = ax_i + b$ on all estimators discussed above. Note that for non-linear transformations (e.g. $y_i = x_i^2$ or $y_i = \ln{x_i})$, no straightforward formulas exist to calculate summary statistics. These then need to be recalculated from scratch from the transformed dataset.

The most commonly used linear transformation is called a **standardization**, which can be written as:
$$
y_i = a x_i + b = \frac{x_i}{s} - \frac{x}{s} = \frac{x_i - x}{s} = z_i.
$$
The standardization expresses how many standard deviations an observation $x_i$ is from the sample mean, since $x_i = x + z_is$. For any standardized variable $z$, the mean is zero and the variance is one.
### 3.6.1. Measures of central tendency or location

- The **median** undergoes an identical linear transformation as the original dataset. As the relative ordering all values does not change for any linear transformation (barring an inversion for $a < 0$), the median of the transformed data set is simply:
$$
M'_e = aM_e + b.
$$

- The **mode** also undergoes an identical linear transformation as the original dataset. It is trivial to see that the relative frequency of observations must be kept under a linear transformation. Therefore, 
$$
M'_o = aM_o + b.
$$

- The **arithmetic mean** also undergoes an identical linear transformation as the original dataset:
$$
y = \frac{1}{n} \sum_{i=1}^{n} y_i
= \frac{1}{n} \sum_{i=1}^{n} ax_i + b
= \frac{1}{n} \left\{ nb +  a \sum_{i=1}^{n} x_i \right\}
= b + a \frac{1}{n} \sum_{i=1}^{n} x_i
= a x + b.
$$

- It is not possible to relate **geometric mean** of a linear transformation to the geometric mean of the original:
$$
G' = \left[ \prod_{i=1}^{n} y_i \right]^{\frac{1}{n}}
= \left[ \prod_{i=1}^{n} ax_i + b \right]^{\frac{1}{n}}
= \left[ b^n + a^n \prod_{i=1}^{n} x_i \right]^{\frac{1}{n}}.
$$  

### 3.6.2. Measures of relative location

- The **$p$-th quantile** undergoes an identical linear transformation as the original dataset. To calculate $c_p$, we first calculate $q = p(1+n)$. For integer values of $q$, 
$$ c_p = y_q = a x_r + b.$$

For non-integer $p$, we round down $q$ down to the nearest integer $r$. Then,
$$
\begin{align}
c'_p 
&= (1-f) y_r + f y_r \nonumber\\
&= y_r + f \left( y_{r+1} - y_r \right) \nonumber\\
&= a x_r + b + f \left( a x_{r+1} + b - \left(a x_r + b \right) \right) \nonumber\\
&= a x_r + b + af\left(x_{r+1} - x_r \right) \nonumber\\
&= a\left[ x_r + f\left(x_{r+1} - x_r \right) \right] + b \nonumber\\
&= a c_p + b. \nonumber
\end{align}
$$

### 3.6.3. Measures of variation or spread

- The **range** neglects the offset of the linear transformation of the original dataset:
$$
R' = y_\mathrm{max} - y_\mathrm{min} 
= a x_\mathrm{max} + b - \left( a x_\mathrm{min} + b \right) 
= a\left( x_\mathrm{max} - x_\mathrm{min} \right) 
= aR.
$$

- The **interquartile range** also neglects the offset of the linear transformation of the original dataset. From the definition of the linearly transformed $p$-th quartile $c_p$:
$$
Q' = Q'_3 - Q'_1
= c'_{0.75} - c'_{0.25}
= a c_{0.75} + b - \left( a c_{0.25} + b \right) 
= a \left( c_{0.75} - c_{0.25}\right) 
= aQ
$$

- The **mean absolute deviation** also neglects the offset of the linear transformation of the original dataset. From the definition of the linearly transformed arithmetic mean $x$:
$$
\mathrm{MAD}' = \frac{1}{n} \sum_{i=1}^{n} \left( y_i - y \right)
= \frac{1}{n} \sum_{i=1}^{n} \left( ax_i + b - \left( ax + b \right) \right)
= \frac{a}{n} \sum_{i=1}^{n} \left( x_i - x \right)
= a\mathrm{MAD}
$$

- The **variance** of a linearly transformed dataset is equal to the original variance times the squared slope:
$$
s'^2 = \frac{1}{n-1} \sum_{i=1}^{n} \left( y_i - y\right)^2
= \frac{1}{n-1} \sum_{i=1}^{n} \left( ax_i + b - \left(ax + b \right) \right)^2
= \frac{a^2}{n-1} \sum_{i=1}^{n} \left( x_i - x\right)^2
= a^2 s^2.
$$

- The **standard deviation** also neglects the offset of the linear transformation of the original dataset. From the definition of the variance $s^2$:
$$
s' = \sqrt{s'^2} = \sqrt{a^2 s^2} = as.
$$

- It is not possible to relate **coefficient of variation** of a linear transformation to the $\mathrm{CV}$ of the original. From the definition of the linearly transformed standard deviation and arithmetic mean:
$$
\mathrm{CV}' = \frac{s'}{y} = \frac{as}{ax + b} = \frac{s}{x + \frac{b}{a}}.
$$

### 3.6.4. Measures of skewness
- **Pearson's skewness coefficient** remains unchanged under linear transformations of the original dataset:
$$
S_P' = \frac{3}{s'} \left(x' - M_e'\right)
= \frac{3}{as} \left(ax + b - \left(aM_e + b\right) \right)
= \frac{3}{s} \left(x - M_e\right)
= S_P.
$$

- **Fisher's skewness coefficient** also remains unchanged under linear transformations of the original dataset:

$$\begin{align}
S_F' &= \frac{n^2}{(n-1)(n-2)} \sum_{i=1}^{n} \left( \frac{y_i - y}{s'} \right)^3\nonumber\\
     &= \frac{n^2}{(n-1)(n-2)} \sum_{i=1}^{n} \left( \frac{ax_i + b - \left(ax + b\right) }{as} \right)^3\nonumber\\
     &= \frac{n^2}{(n-1)(n-2)} \sum_{i=1}^{n} \left( \frac{ax_i - ax }{as} \right)^3\nonumber\\
     &= \frac{n^2}{(n-1)(n-2)} \sum_{i=1}^{n} \left( \frac{x_i - x}{s} \right)^3\nonumber\\
     &= S_F.\nonumber
\end{align}$$

### 3.6.5. Kurtosis

- The **kurtosis** also remains unchanged under linear transformations of the original dataset:

$$\begin{align}
g' &= \frac{n (n+1)}{(n-1)(n-2)(n-3)} \sum_{i=1}^{n} \left( \frac{y_i - y}{s'} \right)^4 - \frac{3 (n-1)^2}{(n-2)(n-3)}\nonumber\\
   &= N_1 \sum_{i=1}^{n} \left( \frac{y_i - y}{s'} \right)^4 - N_2\nonumber\\
   &= N_1 \sum_{i=1}^{n} \left( \frac{ax_i + b - \left(ax + b\right)}{as} \right)^4 - N_2\nonumber\\
   &= N_1 \sum_{i=1}^{n} \left( \frac{ax_i - ax}{as} \right)^4 - N_2\nonumber\\
   &= g.\nonumber
\end{align}$$

## 3.7. Box plots

The **box plot** is a frequently used graphical representation of **univariate ordinal or quantitative data**. It offers the following insights:

- The central part of the data is represented by a box bounded by the **first and third quartile**.
- The central estimates **mean and median** are represented by two differently formatted horizontal lines.
- **Extreme values** are indicated by dots outside of the central box. Typically, an observation $x_i$ is not deemed extreme if:
$$
Q_1 - 1.5Q < x_i < Q_3 + 1.5Q.
$$

```python
taxis = sns.load_dataset('taxis')
minimum_distance = 1
f, a = plt.subplots()
sns.boxplot(data=taxis[taxis['distance'] >= minimum_distance], x='distance', y='pickup_borough', fliersize=1, ax=a)
a.set(xlabel='Distance [km]', ylabel='Pickup borough',title=f'Box plot of taxi rides longer than {minimum_distance} km')
```

## 3.8. Variability charts

>   Uncontrolled variation is the enemy of quality.
>
> -- <cite>W. Edwards Deming<cite>

When improving quality with **Six Sigma**, it is important to first identify the source(s) of variation in the process. To this end, graphics liket he variability charts can be tremendously helpful. In its simplest form, a variability chart can be considered as a **systematic way to create several box plots side by side**. In this example, we will attempt to find the airlines that suffer from most delays when departing to the most common destination airport, based on a sample of more than 5 million flights.

```python
# Constants.
DELAY_QUANTILE = 0.95

# Inferring dtypes for faster reading. Include NaN support "DEPARTURE_DELAY" column.
dtypes = {
    "YEAR": 'int', "MONTH": 'int', "DAY": 'int',
    "AIRLINE": 'category', "FLIGHT_NUMBER": 'int', "TAIL_NUMBER": 'category', 
    "ORIGIN_AIRPORT": 'category', "DESTINATION_AIRPORT": 'category',
    "SCHEDULED_DEPARTURE": 'str', "DEPARTURE_TIME": 'str', "DEPARTURE_DELAY": pd.Int16Dtype()}

# Parsing data in multiple columns (SLOW).
def date_parser(*cols):
    dt = ''.join([*cols])
    return pd.to_datetime(dt, errors="coerce", cache=True, format="%Y %m %d %H%M")

# Data reading and cleaning.
flights = pd.read_parquet(
    'data/flights.parquet', columns = dtypes.keys())
flights = flights.dropna(axis=0, how='any')
flights.columns = [col.lower() for col in flights.columns]

# Seting categorical variables.
flights[["airline", "origin_airport", "destination_airport"]] = flights[
        ["airline", "origin_airport","destination_airport"]].astype("category")

# Setting delay as timedelta and dropping rows with invalid times.
# GOOD IDEA BUT NOT PROPERLY INTEGRATED IN PLOTTING.
# flights["departure_delay"] = pd.to_timedelta(flights["departure_delay"], unit='minutes')

# Setting departures as datetime and removing original columns without formatting.
for old_col, new_col in zip(
    ("scheduled_departure", "departure_time"),
    ("departure_datetime_scheduled", "departure_datetime_actual"),):
    flights["hour"] = flights[old_col].str.slice(stop=2)
    flights["minute"] = flights[old_col].str.slice(start=2)
    flights[new_col] = pd.to_datetime(flights[["year", "month", "day", "hour", "minute"]])
flights = flights.drop(["year", "month", "day", 
                        "hour", "minute",
                        "scheduled_departure", "departure_time"], axis=1)

# Finding the worst flights to the most common destination.
most_common_destination = flights["destination_airport"].mode().item()
most_delayed_flights_to_most_common_destination = flights[
    ( flights["destination_airport"] == most_common_destination )
    &
    ( flights['departure_delay'] > flights['departure_delay'].quantile(DELAY_QUANTILE) )
]
# We note there appears to be a correlation between the airline and the mean departure delay.
# NaT reflects here that no flights left from that airport with sufficient delay.
print(most_delayed_flights_to_most_common_destination.groupby("airline")["departure_delay"].median())

# Plotting boxplots of all flight delays by airline.
f, a = plt.subplots()
sns.boxplot(data=most_delayed_flights_to_most_common_destination, 
            x="airline", y="departure_delay", ax=a)
a.set(xlabel="Airline", ylabel="Departure delay [min]", 
      title=f"Flights with departure delays $> c_{{{DELAY_QUANTILE}}}$ landing in {most_common_destination} by airline")
```

## 3.9. Bivariate data

In a bivariate dataset, we have $n$ observations of both the first variable $x$ and $n$ corresponding observations of the second variable $y$. As an example, both height $x$ and weight $y$ of $n$ people could be measured. Measuring and interpreting the relationship or association between two or more variables is one of the main tasks of statistics. For that purpose, the **covariance** and **covariation** are frequently used to measure the strength of a linear relationship between two quantitative variables.

### 3.9.1. Covariance

The **sample covariance** between variables $x$ and $y$ is defined as:
$$
s_{xy} = \frac{1}{n-1} \sum_{i=1}{n}\left( x_i - x\right) \left( y_i - y \right).
$$

This definition can also be extended towards a full population of $N$ elements, leading to the **population covariance**, described as:
$$
\sigma_{xy} = \frac{1}{N} \sum_{i=1}^{n} \left( x_i - \mu_X \right) \left( y_i - \mu_Y \right),
$$

with $\mu_X$ and $\mu_Y$ the population means of variables $x$ and $y$, respectively. Based on these definitions, it should be clear that **variance** and is simply a special case of **covariance**, namely of the sample $x$ with itself. Therefore, we have $s^2_x = s_{xx}$ and $\sigma^2_x = \sigma_{xx}$. For that reason, covariance must be a number between $-|s_xs_y| < s_{xy} < |s_xs_y|$.

When we examine individual terms $t = \left( x_i - x\right) \left( y_i - y \right)$ in the summation, we notice that:

- $t > 0$ if both $x_i < x$ and $y_i < y$ or if both $x_i > x$ and $y_i > y$. In other words, $x_i$ and $y_i$ grow or shrink with the same tendency.
- $t < 0$ if both $x_i < x$ and $y_i > y$ or if both $x_i > x$ and $y_i < y$. In other words, $x_i$ and $y_i$ grow or shrink with the opposite tendency.

If the number of positive terms $t$ is dominant, we speak of **positive covariance** i.e., the variables $x$ and $y$ change together in the same way (for example, car weight and length). Conversely, if there are many more negative terms $t$, we speak of **negative covariance** where the variables $x$ and $y$ change in opposite ways (for example car weight and miles per gallon). In the scenario that both kinds of terms are present with roughly the same frequency, there is **zero covariance** $x$ and $y$ (for example car brand and the number of friends the drivers has[^1]). Note however, that this does **not** imply that these variables are **independent**. This can be trivially proven by the following example.

Let $x$ be a random variable that is $-1$ or $1$ with probability $0.5$. Then, let $y$ bet a random variables such that $y = 0$ if $x=-1$ and $y$ os randomly $-1$ or $1$ with probability $0.5$ if $x=1$. Clearly, $x$ and $y$ are highly dependent, as knowing $y$ allows a perfect prediction of $x$. Knowing that both their means are $0$, let us examine their covariance:

$$\begin{align}
s_{xy}  =& \sum_{i=1}^{n} f_i \left( x_i - x\right) \left( y_i - y \right)\nonumber\\
        =& \left( -1 - 0 \right) \left( 0 - 0 \right) P(x=-1) \nonumber\\
         &+ \left( 1 - 0 \right) \left( 1 - 0 \right) P(x=1, y=1) \nonumber\\
         &+ \left( 1 - 0 \right) \left( -1 - 0 \right) P(x=1, y=-1) \nonumber\\
        =& \text{ }0\nonumber.
\end{align}
$$

Finally, the downside of covariance is that it is very sensitive to the measurement unit. Therefore, it is difficult to accurately interpret correlations. Luckily, another estimator does not suffer from this problem.

[^1]: It should be noted that a *significant* dip was detected for Audi drivers, but this was assumed to be an outlier.

### 3.9.2. Correlation

**Pearson's correlation coefficient** of the observations $(x_1, y_1), (x_2, y2), ..., (x_n, y_n)$ is defined as:

$$
r_{xy} = \frac{\mathcal{COV}(x, y)}{\mathcal{VAR}(x)\mathcal{VAR}(y)} = \frac{s_{xy}}{s_xs_y} = \frac{\sum_{i=1}^n (x_i - x)(y_i - y)}{\sqrt{\sum_{i=1}^n (x_i - x)^2} \sqrt{\sum_{i=1}^n (y_i - y)^2}}
$$

The correlation coefficient is a number $-1 < r_{xy} < 1$. In case $x$ and $y$ are perfectly positively linearly related, $r_{xy} = 1$ i.e., a change in $x$ leads to a similar change in $y$. Conversely, if $r_{xy} = -1$, $x$ and $y$ are negatively linearly related i.e., a change in $x$ leads to an opposite change in $y$. Finally, in case $r_{xy} = 0$, variables $x$ and $y$ are uncorrelated, meaning there is no linear relationship between them. Note, however, that there still might be a non-linear relationship (quadratic, cubic, logarthmic...) between them.

Below, a **correlation matrix** is plotted for several attributes of more than 200 cars. From this matrix, we can describe some intuitive and less intuitive correlations:

- Width, length and weight of cars are strongly postively correlated.
- Height is only weakly positively related to the above.
- Weight, engine size and horsepower are closely positively correlated, as larger, heavier engines are more powerful.
- City and highway miles-per-gallon are perfectly positively correlated, meaning that ecological cars are more ecological in all urban terrain.
- Logically, both the city and highway miles-per-gallon are strongly negatively correlated with width, height, length and weight.
- Interestingly, expensive cars are not ecological at all. Looking at the $c_{0.99}$ of prices, the three listed cars are 2 Buick oldtimers and a large BMW SUV. Their mean mpg is just over half of that of all other cars.

```python
# Pandas only plots the correlation of numeric columns so we extract these for easy tick labelling.
cars = pd.read_csv('data/cars.csv', index_col=0)
cars_numeric = cars.select_dtypes(include=['int16', 'int32', 'int64', 'float16', 'float32', 'float64'])
corr = cars_numeric.corr()

# Constructing an upper triangle mask of Trues to set these to nan in the correlation matrix.
mask = np.zeros_like(corr, dtype=bool)
mask[np.triu_indices_from(mask)] = True
corr[mask] = np.nan

# Plotting the matrix with a diverging colormap to clearly distinguish positive and negative correlation.
f, a = plt.subplots()
a.matshow(corr, cmap='coolwarm')

# Annotating each lower triangle square
for (y, x), z in np.ndenumerate(corr):
    if not np.isnan(z):
        a.text(x, y, f"{z:.2f}", ha="center", va="center", fontsize="xx-small")

# Decorating axes and title.
a.set_xticks(range(len(cars_numeric.columns)))
a.set_xticklabels(cars_numeric.columns, rotation=90, ha='center')
a.set_yticks(range(len(cars_numeric.columns)))
a.set_yticklabels(cars_numeric.columns, rotation=0, ha='right')
a.set(title=f"Correlation matrix of {len(cars_numeric.columns)} attributes of {len(cars_numeric)} cars")

# Adding a white grid on the minor ticks. Remove the minor ticks from the axes.
a.set_xticks(a.get_xticks() - 0.5, minor=True)
a.set_yticks(a.get_yticks() - 0.5, minor=True)
a.grid(which="minor", color=(1,1,1), linewidth=2)
a.tick_params(which="minor", bottom=False, top=False, left=False)

# Displaying the 1% most expensive cars.
most_expensive = cars[cars["price"] > cars["price"].quantile(0.99)]
print("ALL CARS\n", cars[["citympg", "highwaympg"]].mean())
print("1% MOST EXPENSIVE CARS\n", most_expensive[["citympg", "highwaympg"]].mean())
```

### 3.9.3. Rank correlation

A major downside of **Pearson's correlation coefficient** is its **inability to quantify non-linear relationships** between variables. For such relationships, one can use **Spearman's rank coefficient $S_S$**, which can quantify any **monotone** relationship. A relationship $f(x)$ is monotone if it either preserves or reverses the given order of $x$, but not both within the same domain. In other words, any relationship where $f(x)$ never decreases (or increases) for increasing $x$ is monotone.

During the calculation, the only difference is that Spearman's correlation coeffient relies on ranked data rather than the original dataset. Suppose we try to quantify a relationship between the exam results of English and mathematics for 10 students. First, we rank each student's score from high to low by subject (with 1 associated to the highest score). In case of ties, we can give each observation the average rank. Then, we simply calculate **Pearson's correlation coefficient on the ranked dataset**. Note the difference in the below dataset:

- $S_P = 0.800539$
- $S_S = 0.668696.$

```python
# Creating socres for 10 students.
scores = pd.DataFrame.from_dict({
    'score_english': [56, 75, 45, 71, 61, 64, 58, 80, 76, 61],
    'score_mathematics': [66, 70, 40, 60, 65, 56, 59, 77, 67, 63]
})

# Ranking both scores in descending order.
scores[["rank_english", "rank_mathematics"]] = scores.rank(axis=0, method="average", ascending=False)
corr_scores = scores[[c for c in scores.columns if c.startswith("score")]].corr()
corr_ranks = scores[[c for c in scores.columns if c.startswith("rank")]].corr()

print(f'PEARSON {corr_scores} \n\nSPEARMAN{corr_ranks}')
```

## 3.10 Complementarity of statistics and graphics

Often, a graphical representation of the data provides important insights that go unnoticed when only means, variances and correlations are calculated. Here, we describe 6 cases where the naively calculated correlation coefficient $r_{xy}$ might be misleading.

1. **Outliers can cause correlation.** A blob of randomly distributed data is shown near the origin, combined with 1 outlier in the top right. Only when neglecting this outlier we can come to the correct conclusion that $x$ and $y$ are, indeed, not correlated at all.
2. **Outliers can hide correlation.** Except for one point, the variables $x$ and $y$ are nearly perfectly positively correlated. However, this individual outlier causes a significant drop in the correlation coefficient.
3. **Distinct subpopulations can cause correlation.** Here, two distinct random distributions are plotted in the bottom left and the top right. Clearly, both $x$ and $y$ of both subpopulations are non-correlated, as indicated by the individual correlation coefficients $S_{xy,1}$ and $S_{xy,2}$. However, the correlation coefficient of the total population, $S_{xy}$ is significantly higher.
4. **Distinct subpopulations can hide correlation.** This dataset contains two distinct sets of points that lie more or less on a straight line. Cleary, those two sets must correspond to clearly identifiable groups of objects, for example mean and women. The correlation of the total population, however, is much lower than those of the subpopulations individually.
5. **Non-linear relationships cannot be quantified.** In this case, Pearson's correlation coefficient $S_{xy}$ is unable to identify an obvious quadratic relation between $x$ and $y$. However, when squaring $x$ as well, $S_{x^2y}$ can represent the present correlation.
6. **Correlations are nonsensical if one variable is a component of another.** Here, the any observation $y_i$ is always lower than the corresponding $x_i$. This can be an indication that $y$ is a component of $x$. For example, if $x$ are the total monthly expenses of a household, $y$ might be rent. In that case, $x$ must increase if $y$ increases, leading to a nonsensical (but strong) positive correlation.  

```python
N_SAMPLES = 30
MIN = 0
MAX = 5
NOISE_SCALE = 0.1

def get_corr(df: pd.DataFrame, col1:str='x', col2:str='y', ndigits:int=2) -> float:
    return round(df.corr()[col1][col2], ndigits)


# An outlier can lead to wrong assumptions of presence of correlation correlation in a random dataset.
fake_correlation_outlier = pd.DataFrame(
    np.vstack( [np.random.rand(N_SAMPLES,2), np.array([MAX,5])] ), 
    columns=['x','y'])

fake_correlation_outlier.plot.scatter(x='x', y='y', ax=plt.subplots()[1],
                                      title="1. Outlier causing correlation\n" +
                                            f"$S'_{{xy}} = {get_corr(fake_correlation_outlier)}$ vs " + 
                                            f"$S_{{xy}} = {get_corr(fake_correlation_outlier[:-1])}$")

# An outlier can lead to wrong assumptions about absence of correlation in an ordered dataset.
true_correlation_outlier = pd.DataFrame(
    np.vstack(
    [np.linspace((MIN,MIN), (MAX - NOISE_SCALE, MAX - NOISE_SCALE), N_SAMPLES) +  np.random.rand(N_SAMPLES,2) * NOISE_SCALE, #(x1, y1), (xn, yn), n
     np.array([MAX, 5*MAX])]),
    columns=['x','y']
)

true_correlation_outlier.plot.scatter(x='x', y='y', ax=plt.subplots()[1],
                                      title="2. Outlier hiding correlation\n" +
                                      f"$S'_{{xy}} = {get_corr(true_correlation_outlier)}$ vs " + 
                                      f"$S_{{xy}} = {get_corr(true_correlation_outlier[:-1])}$")

# Two uncorrelated populates can lead to an correlated dataset.
two_groups = pd.DataFrame(
    np.vstack([
        np.random.rand(N_SAMPLES, 2),
        np.random.rand(N_SAMPLES, 2) + MAX
    ]),
    columns=['x','y']
)

two_groups.plot.scatter(x='x', y='y', ax=plt.subplots()[1],
                                      title="3. Subpopulations causing correlation\n" +
                                      f"$S'_{{xy}} = {get_corr(two_groups)}$ vs " + 
                                      f"$S_{{xy, 1}} = {get_corr(two_groups[:N_SAMPLES])}, S_{{xy, 2}}={get_corr(two_groups[N_SAMPLES:])}$")

# Two correlated subpopulations can lead to an uncorrelated total population.
x1 = np.linspace(MIN, MAX, N_SAMPLES) + np.random.rand(N_SAMPLES)
y1 = np.random.rand(1) * MAX * x1 + np.random.rand(1) * MAX/2

x2 = np.linspace(MIN, MAX, N_SAMPLES) + np.random.rand(N_SAMPLES)
y2 = np.random.rand(1) *MAX * x2 + np.random.rand(1) * MAX/2

two_linears = pd.DataFrame(
    np.hstack(
    [np.vstack([x1, y1]),
     np.vstack([x2,y2])]
    ).T,
    columns=['x','y']
)

two_linears.plot.scatter(x='x', y='y', ax=plt.subplots()[1],
                        title="4. Subpopulations hiding correlation\n" +
                        f"$S'_{{xy}} = {get_corr(two_linears)}$ vs " + 
                        f"$S_{{xy, 1}} = {get_corr(two_linears[:N_SAMPLES])}, S_{{xy, 2}}={get_corr(two_linears[N_SAMPLES:])}$")

# Non-linear relationship is still correlated.
x = np.linspace(-MAX, MAX, 2*N_SAMPLES) + np.random.rand(2*N_SAMPLES)
y = x**2 + np.random.rand(2*N_SAMPLES)
nonlinear = pd.DataFrame.from_dict({'x': x, 'y':y})
nonlinear['x_squared'] = nonlinear['x']**2

nonlinear.plot.scatter(x='x', y='y', ax=plt.subplots()[1],
                       title="5. Non-linear relationships\n" +
                       f"$S'_{{xy}} = {get_corr(nonlinear)}$ vs " + 
                       f"$S_{{x^2y}} = {get_corr(nonlinear, col1='x_squared')}$")

# Variable y is a component of x.
xs = np.random.rand(N_SAMPLES, 1) * MAX
ys = xs - np.random.rand(N_SAMPLES, 1) * MAX/2
component_of_other = pd.DataFrame(
    np.hstack([xs, ys]),
    columns=['x','y']
)

component_of_other.plot.scatter(x='x', y='y', ax=plt.subplots()[1], ylim=0,
                                      title="6. $y$ component of $x$\n" +
                                      f"$S'_{{xy}} = {get_corr(component_of_other)}$")
```
